```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

class ClimateAnalyzer:
    def __init__(self, variable):
        self.variable = None
        self.file_path = fr'..\resources\{variable}\complete_dataset_{variable}.csv'
        self.data = None
        self.monthly_avg = None

    def import_data(self):
        """Imports the CSV file and creates a proper DATE column."""
        self.data = pd.read_csv(self.file_path)
        self.variable = self.data.columns[-1]
        self.data["DATE"] = pd.to_datetime(
            self.data["YEAR"].astype(str) + self.data["DOY"].astype(str),
            format="%Y%j"
        )
        
        return self.data

    def create_monthly_average(self):
        """Computes monthly averages of the variable."""
        if self.data is None:
            raise ValueError("Data not loaded. Run import_data() first.")
        
        self.monthly_avg = (
            self.data.groupby(
                [self.data.DATE.dt.month.rename("MONTH"),
                 self.data.DATE.dt.year.rename("YEAR")]
            )
            .mean()[[self.variable]]
            .reset_index()
            .drop(columns=["YEAR"])
        )
        return self.monthly_avg

    def plot_violin(self):
        color_map = {
            'ALLSKY_SFC_UVA': 'darkred',
            'ALLSKY_SFC_UVB': 'blue',
            'CHLOR': 'green',
            'PRECTOTCORR': 'darkblue',
            'RH2M': 'brown',
            'SST': 'gray',
            'T2M': 'red',
            'T2M_MAX': 'orangered',
            'T2M_MIN': 'darkred',
            'WD2M': 'navy',
            'WS2M': 'teal'
        }
        """Plots a violin plot of monthly averages."""
        if self.monthly_avg is None:
            raise ValueError("Monthly average not computed. Run create_monthly_average() first.")
        
        plt.figure(figsize=(8,5))
        sns.violinplot(x="MONTH", y=self.variable, data=self.monthly_avg, inner="quartile", color = color_map[self.variable])
        plt.title(f"{self.variable}")
        plt.xlabel("Month")
        plt.ylabel("Value")
        plt.show()

    def run_pipeline(self):
        """Runs all steps: import → monthly average → plot."""
        self.import_data()
        self.create_monthly_average()
        self.plot_violin()
```

```python
variables = ['ALLSKY_SFC_UVA',
 'ALLSKY_SFC_UVB',
 'CHLOR_M',
 'CHLOR_W',
 'PRECTOTCORR',
 'RH2M',
 'SST_M',
 'SST_W',
 'T2M',
 'T2M_MAX',
 'T2M_MIN',
 'WD2M',
 'WS2M']

for var in variables:
    climate = ClimateAnalyzer(var)
    climate.run_pipeline()
```
